```python
import numpy as np
import pandas as pd
df=pd.read_csv('Countries.csv')
df.head(3)
```

```python
#which country has highest population
df[df['population']==df['population'].max()][['country','population']]
```

```python
#capital of country with highest population
df[df['population']==df['population'].max()]['capital_city']
```

```python
#give top5 countries with highest democratic score
df.nlargest(5,'democracy_score')['country']
```

```python
df.sort_values(by='democracy_score',ascending=False,inplace=True)
df['country'].head(2)
```

```python
#how many total regions are there
df['region'].value_counts().count()
```

```python
#how many countries lie in eastern europe region
df[df['region']=='Eastern Europe']['country']
```

```python
#who is political leader of 2nd highest populated country
df[df['population']==df['population'].nlargest(2).iloc[1]]['political_leader']
```

```python
#how many countries whose political leaders are unknown
df[df['political_leader'].isna()]['country'].count()
```

```python
#how many country have republic in their full name
count=0
def counting(txt):
    global count
    if 'republic' in txt.lower():
        count+=1
    return txt
df['country_long']=df['country_long'].apply(counting)
print(count)
```

```python
#which country in african region has highest population
african_df=df[df['continent']=='Africa']
african_df[african_df['population']==african_df['population'].max()]['country']
```
